## 1. Background Information

### 1.1 Data Context / Introduction to Hotel Reviews

The raw data is a dataset of 515k Hotel Reviews in Europe given by customers as well
as ratings of 1493 luxury hotels extracted from the popular travel and lodging
reservations website : Booking.com
###### How was this data created? 
- The answer is through web scraping. Web scraping is the act of gathering specific data and copying it from the web into a database or spreadsheet for purposes of data analysis. To explain further, all of the information and variables in this dataset are located somewhere on the web page of reviews and is compiled into distinct columns of information. Jason Liu, a Kaggle enthusiast and current Data Analyst at Optiver trading firm is credited for creating this hotel data.

###### What is a hotel review anyway?
- Individuals that are planning their next ideal vacation really do need some advice, and in particular they would like to know what to expect when they begin their search for for accomodation. Because they have only seen pictures, heard briefly about a hotel they are considering, or simply heard propaganda coming from the hotel website itself, they would like to know from previous customers like themselves of specific details they mentioned during their stay and how it compared to their expectations.

###### How hotel reviews are written/verified through Booking.com?
It is possible to write hotel reviews on Booking.com through the following steps:
- (1) Booking a trip through Booking.com 
- (2) Checking in at the property and completing your stay
- (3) A request from Booking.com asking guests about their stay. Through this method, authenticity of the guest’s stay is verified before adding a review to the website.




### 1.2 Discussion of Variable Information and Importance

The data contains 17 variables which describe information about the hotel and reviewer in some of the following ways.
- Background of the reviewer’s history on Booking.com / demographics
- Text of an individual’s review and other attributes related to their review
- Location details & a reviewer’s review history of hotels

###### Hotel Address
- This variable lists the exact location of the property and contains information such as the unit of the property, street name, city, country, and postal code.This is extracted through web scraping in which the address is provided under the name of the property on its main page.

###### Hotel Name
- This is very self-explanatory. The full name of a listing belongs in this variable

###### Lat((Latitude of hotel)
- The term latitude refers to the north-south position of a point on the surface of the earth. It is an angle measure that corresponds to 0° at the equator and 90° at the north and south poles. We indicated if a point is below or above the equator with S (South) or N (North).
- Example: Germany’s Berlin lies more east of the Prime Meridian than Holland’s Amsterdam. This is why Berlin is located at 13.4050° E, a greater numeric value than 4.9036° E for Amsterdam.

###### Review Date
- This is the date when the reviewer posted the review of the hotel. It gives the month, day, and year of the review.

###### Reviewer Score
- Based on one’s experience relating to value for money, service, cleanliness, and more, a score is assigned to their experience from 0 - 10.
- A higher score is reflective of a more positive experience, and a lower one indicative of a guest that may not want to stay there again.
- Booking.com notes that a rating between 7.0 - 8.0 is “Good”, one that is 8.0 - 9.0 (exclusive) is “Very Good”, and 9.0 inclusive - 10 is “Excellent”.

###### Average Score
- Average score (rating) of a hotel. This number will remain consistent across rows that pertain to the same property.

###### Days Since Review
- The number of days since review is the duration between the review date and the web scraping date of that review.
- The date of web scraping done by the author of this date was 8/3/2017. Example: Jaehyeong Lee rated a Booking.com property on 8/1/2017, two days before the scraping date. The value of “Days Since Review” is : 2.
- Booking.com only allows reviews 2 years old and newer to be factored into a hotel rating and displayed on the page. This is why the largest value for Days Since Review is at most 730 days.
- Reason: Reviews past 2 years may become obsolete and not reflect the overall rating a property now deserves.
- Example: The owner of a hotel changed in 2015, providing better service to its customers. By the year 2017, its old management is no longer reflective of the reputation of the hotel.

###### Reviewer Nationality
- The nationality of a reviewer describes the country that they are coming from / a citizen of. Our Data consists of hotels in Paris, France but the nationality of reviewers that traveled to this city are from so many countries in the world.

###### Total Number of Reviews
- This numeric variable indicates the number of valid reviews a property has on Booking.com during scraping.
- A hotel that has been in business since 2010 probably has more total reviews than one that opened in 2016.

###### Total Number of Reviews a Reviewer has Given
- Number of reviews that a reviewer has given in the past on Booking.com
- Reviewers that are well traveled and experienced with Booking.com may have more review counts.
- Example : Web scrapers look at hotel reviews and identify a review of an individual and they make a note of their total # of reviews for all hotels.

######  Negative Review / Review Total Negative Word Counts & Positive Review / Review Total Positive Word Counts
- Text of the negative review is noted in the Negative Review variable, positives are in the Positive Review variable.
- Total amount of words the reviewer wrote to express their negative / positive comment is listed in the corresponding negative / positive word count variable.
- If there was no negative / OR positive comment for an observational unit, the negative review variable will note “No Negative”, “No Positive” respectively.
- Hotels are not perfect. We can expect that for each distinct hotel, there is at least one reviewer noting something negative.
- Fact vs. Opinion
- Some reviews are facts, but the person’s interpretation of the fact is negative. “This hotel is 2 mile from the Eiffel Tower” may or may not be placed in the positive column.
- Other content for reviews consist of opinions. “The staff were very rude to me and my family” is opinion based and negative, placing it in the negative review.

###### Tags
Tags include information regarding 5 aspects related to a reviewer’s status during their trip as well as miscellaneous visit information.
- Length of Stay: The duration in nights of stay at the property before writing the review
- Leisure / Business: The reviewer indicates if the reason for the trip was for work / business, or simply to tour around, see family, etc.
- Status of Travel: Reviewers indicate with whom they traveled : with a pet, a couple, solo travel, etc.
- Type of Room / Bed: Room and Bed type are specified. For example, we may find that one stayed at a “Superior Double Room” or “Double room with Courtyard View”.
- Review Submission from Mobile Device: Reviewers note if they wrote a review via mobile device. If not, there is no note.

###### Additional Number of Scoring
- Some guests made a scoring on the service alone, rather than writing a comprehensive review. This indicates the number of valid scores without reviews.
- Example : A reviewer for a hotel in Paris, France, made a score only for service. He marked scores (rated from 0 - 10) for areas such as WIFI, facilities, and the staff without describing anything in words, followed by receiving his all-around review for service.

###### Observational Unit
- Each row is an observational unit.
- Each row corresponds to a distinct individual’s information relating to their review of a hotel (eg: Review Score, Review Nationality, Positive Review Text). Each row also contains variables regarding the hotel (eg: Hotel name, Hotel Address, Total number of Reviews, Lat,Lng) and values and descriptions in these variables will remain constant throughout all of the reviews for the property.
- Example: Hotel “A” has 100 of its reviews from 100 individuals in this data. This is expanded across 100 rows. This hotel has an average score of 8.4, additional number of scorings of 22, and total number of reviews of 210. These variables all pertain to the property and therefore these values will be repeated across all the 100 rows. However, for example, Review Nationality, Review Score, and Total Number of Reviews Reviewer has will generally change across all the 100 columns as they pertain to the individuals and almost always are not the same.

### 2.3.7 Word Could Image for Positive/Negative reviews.

```python
q = """

    SELECT Positive_Review
    FROM Hotel
    
    WHERE TRIM(Positive_Review) != 'No Positive' AND TRIM(Positive_Review) IS NOT NULL 
    AND TRIM(Positive_Review) NOT IN ('N/A', 'NA', 'None', 'N A', 'na', 'n/a')
    

"""
positives = pd.DataFrame(sqldf(q))
positives.head(5)
q = """

    SELECT Negative_Review
    FROM Hotel
    
    WHERE TRIM(Negative_Review) != 'No Negative' AND TRIM(Negative_Review) IS NOT NULL 
    AND TRIM(Negative_Review) NOT IN ('N/A', 'NA', 'None', 'N A', 'na', 'n/a')

"""


negatives = pd.DataFrame(sqldf(q))
negatives.head(5)
```

```python
import pandas as pd
import numpy as np
import six

def make_to_array(something):
    
    values = something.values
    columns = something.columns
    
    if six.PY2:
        
         types = [(columns[i].encode(), something[j].dtype.type) for (i, j) in enumerate(columns)]
            
    else:
        
          types = [(columns[i], something[j].dtype.type) for (i, j) in enumerate(columns)]  
            
            
    our_types = np.dtype(types)
    
    result = np.zeros(values.shape[0], our_types)
    
    for (i, j) in enumerate(result.dtype.names):
        
        result[j] = values[:, i]
        
    return result

negatives_new = make_to_array(negatives.reset_index())
positives_new = make_to_array(positives.reset_index())
```

## 3. Data Modeling

### 3.1 Anova Models

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import scipy.stats as stats

q = """
        SELECT Reviewer_Score, Continent
        FROM Hotel
        

"""
pre_anova = pd.DataFrame(sqldf(q))
mod = ols('Reviewer_Score ~ Continent',
                data = pre_anova).fit()
            
aov_table = sm.stats.anova_lm(mod, typ = 2)
print(aov_table)
```

When we construct a one-way anova model for analysis of variance between the mean Review Score by Continent, we get a highly significant F value of 260. This indicates that at least one of the pairwise mean comparisons are signficantly different from each other. But we do not know which ones, so we proceed with a Tukey's multiple comparison test.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multicomp import MultiComparison

mc = MultiComparison(pre_anova['Reviewer_Score'], pre_anova['Continent'])
result = mc.tukeyhsd()
 
print(result)
print(mc.groupsunique)
```

The result of the multiple comparison test is that 3 pairwise mean comparisons of Africa to Asia, Africa to Central America, and Europe to Central America are not significantly different because their confidence intervals contain 0 in them. 

### 3.2 Tree based methods for classification

```python
import numpy as np
from sklearn import datasets

# decision tree
from sklearn.tree import DecisionTreeClassifier


Hotel.groupby(['Season','Reason']).count()
```

```python
from sklearn.metrics import confusion_matrix 
from sklearn.cross_validation import train_test_split 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report, confusion_matrix  

Continent_dict = { 'South/Central America':1, 'Europe':2, 
              'Middle East':3, 'North America':4,'Asia':5, 
              'Africa':6 }
Solo_dict = { 'Yes':1, 'No':0 }
Season_dict = { 'Fall':1, 'Winter':2, 'Spring':3, 'Summer':4 }
month_dict = { 'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8,
     'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12 }

new_df = Hotel.loc[:,['Reason','Continent','Solo','Season','month']]
new_df['Continent'] = new_df['Continent'].map(Continent_dict)
new_df['Solo'] = new_df['Solo'].map(Solo_dict)
new_df['Season'] = new_df['Season'].map(Season_dict)
new_df['month'] = new_df['month'].map(month_dict)
filtered_df = new_df[new_df.Reason != 'NOT STATED']

# We removed NOT STATED! 
```

```python
filtered_df.head()
```

Above is the data that we will use for a decision tree model, predicting the variable : Reason. The reason we include these variables is as follows : We first exhibited a relationship between Season containing different proportions of reviewers that traveled for purposes of leisure and business. In addition, when people travel for purposes of leisure, we found they they tend to travel in groups rather than alone. In addition, we saw differences in people that travel for leisure and business between months of the year. In addition, we believe it is worth including the Continent of origin for the reviewer because some nationalities have differences in culture and traveling style and we would like to include this when we want to classify as Leisure of Business. 

```python

def splitdataset(data): 
  
    X = data.values[:, 1:5] 
    Y = data.values[:, 0] 
  
    X_train, X_test, y_train, y_test = train_test_split(  
    X, Y, test_size = 0.3, random_state = 139) 
      
    return X, Y, X_train, X_test, y_train, y_test 
      
def entropy_method(X_train, X_test, y_train): 
  
    entropy_lf = DecisionTreeClassifier( 
            criterion = "entropy", random_state = 139, 
            max_depth = 5, min_samples_leaf = 5) 
  
    entropy_lf.fit(X_train, y_train) 

    return entropy_lf 
  
# Function to make predictions 
def prediction(X_test, lf):   
    y_pred = lf.predict(X_test) 
    return y_pred 
      
def accuracy_results(y_test, y_pred): 
      
    print ("Accuracy:", 
    accuracy_score(y_test,y_pred)*100) 
      
    print("Report : ", 
    classification_report(y_test, y_pred)) 
  
def decision_tree(): 
      
    X, Y, X_train, X_test, y_train, y_test = splitdataset(filtered_df) 
    clf_entropy = entropy_method(X_train, X_test, y_train) 
    
    y_pred_entropy = prediction(X_test, clf_entropy) 
    accuracy_results(y_test, y_pred_entropy) 
    
    return clf_entropy

    
decision_tree()
```

Our Decision tree model predicted the category of reason : business or leisure with 86.25% total accuracy by our confusion matrix! We have done better than guessing. For business, we correctly labeled the reason 62% of the time, but for leisure, we correctly labeled the reason 90% of the time.

### 3.3 KNN Classification

```python
from sklearn.neighbors import KNeighborsClassifier  
y = filtered_df.Reason
X = filtered_df.drop('Reason', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3)

#Here we consider the test size to be 30% and training size to be 70%.

classifiers = KNeighborsClassifier(n_neighbors=5)  
classifiers.fit(X_train, y_train)



y_pred = classifiers.predict(X_test)  

print(confusion_matrix(y_test, y_pred))  
print(classification_report(y_test, y_pred)) 


error_values = []
for i in range(1, 40):  
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred1 = knn.predict(X_test)
    error_values.append(np.mean(pred1 != y_test))

plt.figure(figsize=(12, 6))  
plt.plot(range(1, 40), error_values, color='blue', linestyle='dashed', marker='o',  
         markerfacecolor='red', markersize=10)
plt.title('Error Rate each K Value')  
plt.xlabel('K Value')  
plt.ylabel('Mean Errors')
```

We see from the confusion matrix for KNN classification that the accuracy rate is 85.9 %, and the accuracy rate for Decision Trees is 86.2504%. Therefore, we determine that Tree based methods for this data, predicting leisure or business, is a better classifier than KNN. The value of the MSE changes as you increase K. Once the value of K is 25 or greater, the error rate stabilizes in KNN.
